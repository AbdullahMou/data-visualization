```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set()
```

```python
 crash_rate=sns.load_dataset('car_crashes')
```

```python
sns.relplot(x='speeding', y='total', hue=crash_rate.speeding, data=crash_rate);
```

```python
mean =crash_rate.groupby('speeding', as_index=False).mean()
```

```python
sns.lmplot(x='speeding', y='total', data=crash_rate, line_kws={'color': 'red'})
```

```python
graph = sns.barplot(x='speeding', y='total', data=mean)
graph.set_xticklabels(labels=list("")[1:])
```
