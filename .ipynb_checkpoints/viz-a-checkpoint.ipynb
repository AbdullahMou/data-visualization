```python
import seaborn as sns
sns.set()
titanic= sns.load_dataset('titanic')
First = titanic[titanic['class']=='First']
Second = titanic[titanic['class']=='Second']
Third = titanic[titanic['class']=='Third']
titanic
```

```python
sns.relplot(x='age', y='alive', data=First)
sns.relplot(x='age', y='alive', data=Second)
sns.relplot(x='age', y='alive', data=Third)
```

```python
sns.pairplot(titanic, hue="sex", markers=["o", "D"])
```

**Conclusions**
* median age for surviver is less than the dead
